```python
import matplotlib.pyplot as plt
import random
import time
```

```python
def generate_array(n):
    array = [0] * n
    for i in range(n):
        random_num = random.randint(1, n)
        array[i] = random_num
    return array
```

```python
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1
```

```python
n = [10000000, 50000000, 100000000]
time_array = [0] * len(n)
for i in range(len(n)):
    array = generate_array(n[i])
    print("when array size =", n[i])

    start = time.time() * 1000
    print("start:",start)

    print(linear_search(array, -1))

    end = time.time() * 1000
    print("end:", end)
    print("all operation:", end - start, "ms")

    time_array[i] = end - start
    print()

  # cal the avg
sum_time = 0
avg = 0
for i in range(len(time_array)):
    sum_time += time_array[i]
    avg = sum_time / len(time_array)

print("avg:", avg)
```

```python
plt.plot(n, time_array)
plt.title("Linear Search Performance")
plt.xlabel("Array size ")
plt.ylabel("Time in ms")
plt.show()
```
