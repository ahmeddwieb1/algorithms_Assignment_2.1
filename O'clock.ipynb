```python
def increment(number):
    if number == 3:
        return 0
    else:
        return number + 1
```

```python
letters = ["A", "B", "C", "D", "E", "F", "G", "H", "I"]

with open("input.txt") as f:
    nums = list(map(int, f.read().split()))


values = {letters[i]: nums[i] for i in range(9)}
print (values)
```

```python
def values_to_tuple(values):
    result = []
    # Add the value of each letter to the list in order
    for letter in letters:
        value = values[letter]
        result.append(value)

    return tuple(result)
```

```python
def actions(values):
    actions_list   = []
    for i in values:
        if values[i] != 0:
          actions_list.append(i)
    return actions_list
#test
print(actions(values))
```

```python
moves = {
    1: ["A","B","D","E"],
    2: ["A","B","C"],
    3: ["B","C","E","F"],
    4: ["A","D","G"],
    5: ["B","D","E","F","H"],
    6: ["C","F","I"],
    7: ["D","E","G","H"],
    8: ["G","H","I"],
    9: ["E","F","H","I"],
}
```

```python
def winner(values):
    # Check if all letters have value 0
    for letter in letters:
        if values[letter] != 0:
            return False
    return True
```

```python
def terminal(values):
    active_letters = {l for l in values if values[l] != 0}
    valids = []

    for m in moves:
        if all(letter in active_letters for letter in moves[m]):
            valids.append(m)

    return valids
print(terminal(values))
```

```python
def apply_move(values, move_number):
    new_vals = values.copy()

    for letter in moves[move_number]:
        if new_vals[letter] != 0:
            new_vals[letter] = increment(new_vals[letter])
    return new_vals
```

# monstar level

```python
def dfs_search(start, max_depth=30):

    visited = set()

    def dfs(state, path, depth):
        key = values_to_tuple(state)

        if key in visited:
            return None
        visited.add(key)

        if winner(state):
            return path

        if depth >= max_depth:
            return None

        valids = terminal(state)
        if not valids:
            return None

        for m in valids:
            nxt = apply_move(state, m)
            res = dfs(nxt, path + [m], depth + 1)
            if res is not None:
                return res

        return None

    return dfs(start, [], 0)
```

```python
if __name__ == "__main__":

    with open("input.txt") as f:
        nums = list(map(int, f.read().split()))

    start_values = {letters[i]: nums[i] for i in range(9)}

    solution = dfs_search(start_values, max_depth=30)

    if solution is None:
        print("No solution found.")
        with open("output.txt", "w") as out:
            out.write("")
    else:
        result = "".join(str(x) for x in solution)
        print("Solution:", result)
        with open("output.txt", "w") as out:
            out.write(result)
```
